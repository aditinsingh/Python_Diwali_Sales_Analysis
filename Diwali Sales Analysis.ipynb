# Exploratory Data Analysis

# Gender

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
ax = sns.countplot(x = 'Gender',data = df, hue = 'Gender', palette=custom_palette)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue = 'Gender', palette=custom_palette)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

# Age

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
ax = sns.countplot(x = 'Age Group',data = df, hue = 'Gender', palette=custom_palette)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue = 'Age Group', palette=custom_palette)
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female


# State

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'State',y= 'Orders' ,data = sales_state, hue = 'State', palette=custom_palette)
```

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'State',y= 'Amount' ,data = sales_state, hue = 'State', palette=custom_palette)
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

# Marital Status

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
ax = sns.countplot(data = df, x = 'Marital_Status', hue= 'Marital_Status', palette=custom_palette, legend = False)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_mts = df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status',y= 'Amount' ,data = sales_mts, hue = 'Marital_Status', palette=custom_palette)
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
ax = sns.countplot(data = df, x = 'Occupation', hue ='Occupation', palette=custom_palette)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation',y= 'Amount' ,data = sales_occ, hue = 'Occupation', palette=custom_palette)
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation secto

# Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
ax = sns.countplot(data = df, x = 'Product_Category', hue ='Product_Category', palette=custom_palette)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_pc = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category',y= 'Amount' ,data = sales_pc, hue = 'Product_Category', palette=custom_palette)
```

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_pc = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category',y= 'Amount' ,data = sales_pc, hue = 'Product_Category', palette=custom_palette)
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
custom_palette = {'#eabfff','#3c005a','#d580ff','#800080'}
sales_pid = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_ID',y= 'Orders' ,data = sales_pid, hue = 'Product_ID', palette=custom_palette)
```
